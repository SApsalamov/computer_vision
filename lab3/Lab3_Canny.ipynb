```python
import numpy as np
from skimage import data
import matplotlib.pyplot as plt
```

```python
image = data.camera()
```

```python
plt.imshow(image, cmap=plt.cm.gray)
```

```python
mm = np.mean(image)
mm
```

```python
binary = image > mm
```

```python
plt.imshow(binary, cmap=plt.cm.gray)
```

```python
from skimage import feature
edges1 = feature.canny(image, sigma=2)
```

```python
plt.imshow(edges1, cmap=plt.cm.gray)
```

```python
import random
image = data.camera()
for i in range(image.shape[0]):
    for j in range(image.shape[1]):
        image[i, j] += random.randint(-20, 20)
```

```python
plt.imshow(image, cmap=plt.cm.gray)
```

```python
from skimage import feature

edges1 = feature.canny(image, sigma=4, high_threshold=36, low_threshold=20)
plt.imshow(edges1, cmap=plt.cm.gray)
```

```python
from skimage import feature
for s in range(3,5):
    print(f'sigma: {s}')
    for h in range(0,40):
        for l in range(0,30):
            plt.figure(h)
            #print(f'high_threshold={h}, low_threshold={l}')
            plt.title(f'high_threshold={h}, low_threshold={l}')
            edges1 = feature.canny(image, sigma=s, high_threshold=h, low_threshold=l)
            plt.imshow(edges1, cmap=plt.cm.gray)
```
